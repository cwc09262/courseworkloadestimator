```python
Average_df = df.groupby("Course Code").agg({
    "Year": lambda x: x.mode().iloc[0],
    "Credits": "mean",
    "Has Writing": lambda x: x.mode().iloc[0],
    "Preference": lambda x: x.mode().iloc[0],
    "Aprox Time": "mean"
}).reset_index()

Average_df["Aprox Time"] = Average_df["Aprox Time"].round(0)

print("Average Of All Data For Each Class:")
```
